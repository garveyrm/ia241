# Global COVID-19 Data Analysis

```python
%matplotlib inline
```

```python
import pandas
```

```python
df = pandas.read_excel('s3://garveyrm-ia241/COVID-19-geographic-disbtribution-worldwide-2020-05-01.xlsx')

df
```

# Sum of cases per month

```python
sum_monthly_cases_per_country = df.groupby('dateRep').sum()['cases']

sum_monthly_cases_per_country.plot()
```

There was a small peak in February.

The Global pandemic began in the second half of March into April.

The highest peak occured so far at the end of April

# Sum of deaths per month

```python
sum_monthly_deaths_per_country = df.groupby('dateRep').sum()['deaths']

sum_monthly_deaths_per_country.plot()
```

The most significant spike in deaths was in mid-april.

It is likely the highest peak of death is passed us unless another wave of the virus hits after this one.

# Global Analysis of Cases & Deaths

```python
sum_cases_by_country = df.groupby('continentExp').sum()['cases']
sum_cases_by_country.nlargest(10).plot.bar()
```

```python
sum_death_by_country = df.groupby('continentExp').sum()['deaths']
sum_death_by_country.nlargest(10).plot.bar()
```

Although America has the largest number of cases, Europe exceeds America in the number of deaths by approximately 50 thousand.

# Conclusion

One notable relationship existing in the data is cases and deaths are positively correlated.

Increasing cases continue at a more steady rate than deaths. Deaths reached an all-time high in mid-April and has not peaked like that since then. Monthly case sums continue at a more constant rate.

Even though Europe received the virus before America, data indicates those infected in Europe are much more likely to die than in America. However, this data is susceptible to change.
