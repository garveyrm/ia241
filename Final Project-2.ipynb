```python
import pandas
```

```python
df = pandas.read_excel('s3://garveyrm-ia241/COVID-19-geographic-disbtribution-worldwide-2020-05-01.xlsx')

df
```

# Number of cases per month

```python
sum_monthly_cases_per_country = df.groupby('month').sum()['cases']

df.plot.scatter(x='month',y='cases')
```

# Number of deaths per month

```python
sum_monthly_deaths_per_country = df.groupby('month').sum()['deaths']

sum_monthly_deaths_per_country.plot.bar()
```

Deaths reached an all time high in april.

# Average cases per day

```python
Average_number_cases_per_day = df.groupby('day').mean()['cases']

print(Average_number_cases_per_day)
```

```python
Average_cases_per_day.plot.bar()
```

Cases are steadily increasing with a drop going into May.

# Average deaths per day

```python
Average_number_deaths_per_day = df.groupby('day').mean()['deaths']

print(Average_number_deaths_per_day)
```

```python
Average_number_deaths_per_day.plot.bar()
```

It seems the peeks of death is passed and death rates are decreasing in May.
